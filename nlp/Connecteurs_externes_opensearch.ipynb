# PEPS
    Les descripteurs Opensearch de PEPS sont:
    "https://peps.cnes.fr/resto/api/collections/{collection}/describe.xml"

```python
parameters = {
    'tileid' : "31TCJ",
    'cloudCover' : '[10,70]'
}
collection = "S2ST"
```

```python
def PEPS_descr_url(collection):
    descr_URL = config['PEPS_url']
    if collection in config['PEPS_collections']:
      descr_URL += collection
    descr_URL += "/describe.xml"
    return descr_URL
```

```python
descr_PEPS = PEPS_descr_url(collection)
print(descr_PEPS)
```

```python
from opensearch import Client
PEPS_client = Client(descr_PEPS)
PEPS_results = PEPS_client.search('31TCJ')
# how to search for multiple values?

print("Nr of results: ", len(PEPS_results))
for result in PEPS_results:
    print(result)
    break
```
